```python
import datetime
import pandas as pd
import requests
pd.__version__


league_ID = "1139611"
season_ID = "2018"
curSeason = "2022"  ## used for grabbing previous seasons in the league!

draft_df = getDraftPicks(season_ID, league_ID)  ## a users player they drafted
# draft_df.sample(5)

player_df = getPlayerData(season_ID)  ## a players data like proffessional team they play for
# player_df.sample(5)

team_df = getTeamData(season_ID)  ## Team info like a location and name of team
# team_df.sample(5)

user_df = getUsers(league_ID, season_ID)  ## getting the names of the teams of each user ex. Andrews Team

league_years = get_years(league_ID, curSeason)
league_years




```

```python
df2 = pd.merge(draft_df, player_df, how= "inner", left_on="playerId", right_on="player_id")
df2.sample(5)
```

```python
final_df = pd.merge(df2, team_df, how="inner", left_on="proTeamId", right_on="team_id")
final_df.sample(5)
```

```python
## position mapping for position ID!
##He hard coded this with a map
## right now positions are listed as numbers not wr rb... the fix:

position_mapping = {
    1:'QB',
    2:'RB',
    3:'WR',
    4:'TE',
    5:'K',
    6:'D/ST'
}

league_draft = final_df.replace({"defaultPositionId":position_mapping})
league_draft.sample(5)
```

Mapping Fantasy Names!

```python
## similar to the position mapping, this will make an dynamic mapping 
## based on your leagues names! Custom by Andrew Mina.
user_map = {}
for i in range(user_df.shape[0]):
    user_map[user_df['teamId'][i]] = user_df['abbrev'][i]

user_map


```

```python
league_draft_info = league_draft.replace({"teamId":user_map})
league_draft_info.sample(5)
```

```python
## picking certain data we want and renaming it
league_draft_final = league_draft_info[['overallPickNumber', 'teamId', 'defaultPositionId', 'fullName', 'team name' ]].copy()
league_draft_final.rename(columns = {'overallPickNumber':'Pick', 'teamId':'Fantasy Team', 'defaultPositionId':'Position', 'fullName':'Player', 'team name':'Player Team'}, inplace = True)
league_draft_final.sample(5)
```

```python
all_drafts_df = pd.DataFrame()
```

```python
## repeating all the above processes for every year in the league
## adding a new column for year!

for year in league_years:
    print(year)
    ## getting all the needed info for each year
    draft_df = getDraftPicks(year, league_ID)  ## a users player they drafted
    player_df = getPlayerData(year)  ## a players data like proffessional team they play for
    team_df = getTeamData(year)  ## Team info like a location and name of team

    ## Merging the tables together
    df2 = pd.merge(draft_df, player_df, how = "inner", left_on= "playerId", right_on= "player_id")
    final_df = pd.merge(df2, team_df, how="inner", left_on="proTeamId", right_on="team_id")

    ## adding position names, users team names, and renaming info 
    league_draft = final_df.replace({"defaultPositionId":position_mapping})
    league_draft_info = league_draft.replace({"teamId":user_map})
    league_draft_final = league_draft_info[['overallPickNumber', 'teamId', 'defaultPositionId', 'fullName', 'team name' ]].copy()
    league_draft_final.rename(columns = {'overallPickNumber':'Pick', 'teamId':'Fantasy Team', 'defaultPositionId':'Position', 'fullName':'Player', 'team name':'Player Team'}, inplace = True)

    league_draft_final['year'] = year
    league_draft_final['year'] = league_draft_final['year'].apply(str)
    # all_drafts_df = all_drafts_df.append(league_draft_final)
    pd.concat([all_drafts_df, league_draft_final], ignore_index=True)
all_drafts_df
    
    
```
